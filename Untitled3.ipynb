```python
def preprocess_audio(file_path, target_sr=16000, duration=1):
    """Load, resample to 16kHz, apply noise gate."""
    audio, sr = librosa.load(file_path, sr=target_sr, duration=duration)
    # Zero-pad if shorter than 1 second
    expected_len = target_sr * duration
    if len(audio) < expected_len:
        audio = np.pad(audio, (0, expected_len - len(audio)))
    # Noise gate: suppress frames below energy threshold
    rms = librosa.feature.rms(y=audio)[0]
    threshold = np.percentile(rms, 15)
    hop = 512
    mask = np.repeat(rms > threshold, hop)[: len(audio)]
    audio = audio * mask
    return audio, sr

# Quick test
sample = os.path.join(snoring_path, os.listdir(snoring_path)[0])
a, s = preprocess_audio(sample)
print(f"Sample rate: {s} Hz | Length: {len(a)} samples ({len(a)/s:.2f}s)")
```

```python
def extract_full_features(file_path):
    """
    Returns feature vector: 13 MFCC + 13 delta + 13 delta-delta
                           + centroid + rolloff + ZCR + RMS = 43 features
    """
    audio, sr = preprocess_audio(file_path)

    # --- MFCC (PI2) ---
    mfcc = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
    mfcc_mean = np.mean(mfcc.T, axis=0)           # shape (13,)

    # --- Delta + Delta-Delta (PI4) ---
    delta1 = librosa.feature.delta(mfcc)
    delta2 = librosa.feature.delta(mfcc, order=2)
    d1_mean = np.mean(delta1.T, axis=0)           # shape (13,)
    d2_mean = np.mean(delta2.T, axis=0)           # shape (13,)

    # --- Spectral features (PI3) ---
    centroid  = np.mean(librosa.feature.spectral_centroid(y=audio, sr=sr))
    rolloff   = np.mean(librosa.feature.spectral_rolloff(y=audio, sr=sr))
    bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=audio, sr=sr))
    zcr       = np.mean(librosa.feature.zero_crossing_rate(audio))
    rms       = np.mean(librosa.feature.rms(y=audio))

    features = np.hstack([mfcc_mean, d1_mean, d2_mean,
                          centroid, rolloff, bandwidth, zcr, rms])
    return features   # total: 43 features

# Test
f = extract_full_features(sample)
print(f"Feature vector length: {len(f)}")   # expect 43
```

```python
def calculate_severity(audio, sr):
    rms      = np.mean(librosa.feature.rms(y=audio))
    centroid = np.mean(librosa.feature.spectral_centroid(y=audio, sr=sr))
    rms_score  = min(rms * 1000, 100)
    freq_score = min(centroid / 50, 100)
    return round(min(0.6 * rms_score + 0.4 * freq_score, 100), 2)

def severity_level(score):
    if   score < 20: return "No Snoring"
    elif score < 40: return "Mild"
    elif score < 60: return "Moderate"
    elif score < 80: return "High"
    else:             return "Severe"

def detect_silence(audio, sr, threshold=0.01, min_duration_s=10):
    """
    Energy-based silence detection.
    Returns: list of (start_s, end_s) for silent segments >= min_duration_s.
    """
    hop       = 512
    energy    = librosa.feature.rms(y=audio, hop_length=hop)[0]
    silent    = energy < threshold
    frame_dur = hop / sr

    events, in_silence, start = [], False, 0
    for i, s in enumerate(silent):
        if s and not in_silence:
            in_silence, start = True, i * frame_dur
        elif not s and in_silence:
            dur = i * frame_dur - start
            if dur >= min_duration_s:
                events.append((round(start, 2), round(i * frame_dur, 2)))
            in_silence = False
    return events

print("✅ Severity + silence functions defined")
```

```python
features, labels, severity_list = [], [], []
silence_counts = []

for folder, label in [(snoring_path, 1), (non_snoring_path, 0)]:
    for fname in os.listdir(folder):
        if not fname.lower().endswith(('.wav', '.mp3', '.ogg')):
            continue
        path = os.path.join(folder, fname)
        try:
            audio, sr = preprocess_audio(path)
            feat      = extract_full_features(path)
            sev       = calculate_severity(audio, sr)
            sil       = detect_silence(audio, sr, min_duration_s=0.3)

            features.append(feat)
            labels.append(label)
            severity_list.append(sev)
            silence_counts.append(len(sil))
        except Exception as e:
            print(f"  Skipped {fname}: {e}")

X          = np.array(features)
y_class    = np.array(labels)
y_severity = np.array(severity_list)
y_silence  = np.array(silence_counts)

print(f"Dataset shape : {X.shape}")
print(f"Snoring       : {y_class.sum()} samples")
print(f"Non-snoring   : {(y_class==0).sum()} samples")
print(f"Severity range: {y_severity.min():.1f} – {y_severity.max():.1f}")
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

(X_train, X_test,
 y_train_cls, y_test_cls,
 y_train_sev, y_test_sev) = train_test_split(
    X_scaled, y_class, y_severity,
    test_size=0.2, random_state=42, stratify=y_class
)

print(f"Train: {X_train.shape[0]} | Test: {X_test.shape[0]}")
print(f"Class balance (test): snoring={y_test_cls.sum()}, non={len(y_test_cls)-y_test_cls.sum()}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, regularizers, callbacks
from tensorflow.keras.utils import to_categorical

print(f"TensorFlow version: {tf.__version__}")
```

```python
def augment_audio(audio, sr):
    """Random augmentation: pitch shift OR time stretch OR add noise."""
    choice = np.random.randint(3)
    if choice == 0:
        # Pitch shift ±2 semitones
        steps = np.random.uniform(-2, 2)
        return librosa.effects.pitch_shift(audio, sr=sr, n_steps=steps)
    elif choice == 1:
        # Time stretch 0.8x – 1.2x
        rate = np.random.uniform(0.8, 1.2)
        stretched = librosa.effects.time_stretch(audio, rate=rate)
        # Re-pad or trim to original length
        target = len(audio)
        if len(stretched) > target:
            return stretched[:target]
        return np.pad(stretched, (0, target - len(stretched)))
    else:
        # Add Gaussian noise (SNR ~20dB)
        noise = np.random.randn(len(audio)) * 0.005
        return (audio + noise).astype(np.float32)

print("✅ Augmentation function defined")
```

```python
TARGET_SR  = 16000
AUDIO_LEN  = 16000   # 1 second

raw_audios, raw_labels = [], []

for folder, label in [(snoring_path, 1), (non_snoring_path, 0)]:
    for fname in os.listdir(folder):
        if not fname.lower().endswith(('.wav', '.mp3', '.ogg')):
            continue
        try:
            a, _ = preprocess_audio(os.path.join(folder, fname))
            raw_audios.append(a)
            raw_labels.append(label)
            # Augment snoring samples (class=1) only
            if label == 1:
                aug = augment_audio(a, TARGET_SR)
                raw_audios.append(aug)
                raw_labels.append(label)
        except: pass

X_raw = np.array(raw_audios)[..., np.newaxis]   # (N, 16000, 1)
y_raw = np.array(raw_labels)

X_r_train, X_r_test, y_r_train, y_r_test = train_test_split(
    X_raw, y_raw, test_size=0.2, random_state=42, stratify=y_raw)

print(f"1D CNN data — Train: {X_r_train.shape}, Test: {X_r_test.shape}")
```

```python
def build_1d_cnn(input_length=16000):
    inp = layers.Input(shape=(input_length, 1))

    x = layers.Conv1D(32, 64, strides=4, padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(inp)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling1D(4)(x)

    x = layers.Conv1D(64, 32, padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling1D(4)(x)

    x = layers.Conv1D(128, 16, padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.GlobalAveragePooling1D()(x)

    x = layers.Dense(128, activation='relu',
                     kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(0.4)(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inp, out)
    model.compile(optimizer='adam', loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

cnn1d = build_1d_cnn()
cnn1d.summary()
```

```python
early_stop = callbacks.EarlyStopping(monitor='val_loss', patience=5,
                                     restore_best_weights=True)
lr_sched   = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                                         patience=3, verbose=1)

hist1d = cnn1d.fit(
    X_r_train, y_r_train,
    validation_data=(X_r_test, y_r_test),
    epochs=30, batch_size=32,
    callbacks=[early_stop, lr_sched],
    verbose=1
)

# Evaluate
loss1d, acc1d = cnn1d.evaluate(X_r_test, y_r_test, verbose=0)
y_pred_1d = (cnn1d.predict(X_r_test) > 0.5).astype(int).flatten()
print(f"\n1D CNN — Accuracy: {acc1d:.4f}")
print(f"         F1-Score: {f1_score(y_r_test, y_pred_1d):.4f}")
print(f"         ROC-AUC : {roc_auc_score(y_r_test, cnn1d.predict(X_r_test).flatten()):.4f}")
```

```python
N_MELS, N_FFT, HOP = 64, 1024, 512

def get_melspec(file_path, augment=False):
    audio, sr = preprocess_audio(file_path)
    if augment:
        audio = augment_audio(audio, sr)
    mel = librosa.feature.melspectrogram(
        y=audio, sr=sr, n_mels=N_MELS, n_fft=N_FFT, hop_length=HOP)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    # Normalise to [0,1]
    mel_db = (mel_db - mel_db.min()) / (mel_db.max() - mel_db.min() + 1e-9)
    return mel_db[..., np.newaxis]   # (64, T, 1)

specs, spec_labels = [], []
for folder, label in [(snoring_path, 1), (non_snoring_path, 0)]:
    for fname in os.listdir(folder):
        if not fname.lower().endswith(('.wav', '.mp3', '.ogg')):
            continue
        try:
            specs.append(get_melspec(os.path.join(folder, fname)))
            spec_labels.append(label)
        except: pass

# Pad/crop to uniform width
T = min(s.shape[1] for s in specs)
specs = np.array([s[:, :T, :] for s in specs])
spec_labels = np.array(spec_labels)

Xs_tr, Xs_te, ys_tr, ys_te = train_test_split(
    specs, spec_labels, test_size=0.2, random_state=42, stratify=spec_labels)
print(f"2D CNN data — Train: {Xs_tr.shape}, Test: {Xs_te.shape}")
```

```python
def build_2d_cnn(input_shape):
    inp = layers.Input(shape=input_shape)

    x = layers.Conv2D(32, (3,3), padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(inp)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Conv2D(64, (3,3), padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Dropout(0.2)(x)

    x = layers.Conv2D(128, (3,3), padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.GlobalAveragePooling2D()(x)

    x = layers.Dense(256, activation='relu',
                     kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(0.5)(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    m = models.Model(inp, out)
    m.compile(optimizer='adam', loss='binary_crossentropy',
              metrics=['accuracy'])
    return m

cnn2d = build_2d_cnn(Xs_tr.shape[1:])
cnn2d.summary()
```

```python
hist2d = cnn2d.fit(
    Xs_tr, ys_tr,
    validation_data=(Xs_te, ys_te),
    epochs=30, batch_size=32,
    callbacks=[early_stop, lr_sched], verbose=1
)

loss2d, acc2d = cnn2d.evaluate(Xs_te, ys_te, verbose=0)
y_pred_2d = (cnn2d.predict(Xs_te) > 0.5).astype(int).flatten()
print(f"\n2D CNN — Accuracy: {acc2d:.4f}")
print(f"         F1-Score: {f1_score(ys_te, y_pred_2d):.4f}")

# Training curves — both models side by side
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
for ax, hist, title in zip(axes, [hist1d, hist2d], ["1D CNN", "2D CNN"]):
    ax.plot(hist.history['accuracy'],    label='Train acc')
    ax.plot(hist.history['val_accuracy'], label='Val acc')
    ax.plot(hist.history['loss'],         label='Train loss', ls='--')
    ax.plot(hist.history['val_loss'],     label='Val loss',   ls='--')
    ax.set_title(title); ax.legend(); ax.set_xlabel("Epoch")
plt.tight_layout(); plt.show()
```

```python
def detect_apnea_events(severity_list, threshold=20):
    """Count low-energy segments as proxy apnea events."""
    return sum(1 for s in severity_list if s < threshold)

def calculate_osa_risk(events, severities, recording_hours=8):
    """
    Enhanced risk score:
    - events_per_hour = AHI proxy
    - severity_penalty = mean(severities) / 100
    Returns: (risk_label, ahi_proxy, score, confidence_interval)
    """
    ahi = events / recording_hours
    mean_sev = np.mean(severities) if len(severities) > 0 else 0
    weighted_score = 0.7 * ahi + 0.3 * (mean_sev / 100) * 30

    # Confidence interval (±20% of estimate)
    ci_low  = round(weighted_score * 0.8, 2)
    ci_high = round(weighted_score * 1.2, 2)

    if   ahi < 5:  label = "Low Risk (AHI <5)"
    elif ahi < 15: label = "Moderate Risk (AHI 5–15)"
    elif ahi < 30: label = "High Risk (AHI 15–30)"
    else:           label = "Severe Risk (AHI ≥30)"

    return {
        "risk_label":   label,
        "ahi_proxy":    round(ahi, 2),
        "weighted_score": round(weighted_score, 2),
        "ci":           (ci_low, ci_high),
        "mean_severity": round(mean_sev, 2),
        "events":       events
    }

# Apply to full dataset
apnea_events = detect_apnea_events(y_severity.tolist())
risk_result  = calculate_osa_risk(apnea_events, y_severity.tolist())

print("OSA Risk Assessment")
print("===================")
for k, v in risk_result.items():
    print(f"  {k:20s}: {v}")
```

```python
import streamlit as st
import librosa, librosa.display
import numpy as np
import matplotlib.pyplot as plt
import soundfile as sf
import os, pickle
from sklearn.preprocessing import StandardScaler

# ── Page config ──────────────────────────────────────────────
st.set_page_config(page_title="OSA Snoring Screener",
                   page_icon="🫁", layout="wide")

st.title("🫁 OSA Snoring Detection & Risk Screening")
st.caption("Upload an audio file to analyse snoring severity and OSA risk")

# ── Load saved models (run notebook first to pickle them) ────
# import pickle
# with open('models.pkl','wb') as f:
#     pickle.dump({'svm':svm_model,'reg':reg_model,'scaler':scaler}, f)

# ── Reuse helper functions from notebook ─────────────────────
def preprocess_audio(path, sr=16000):
    a, s = librosa.load(path, sr=sr, duration=1)
    if len(a) < sr: a = np.pad(a, (0, sr - len(a)))
    return a, s

def extract_features(path):
    a, sr = preprocess_audio(path)
    mfcc = librosa.feature.mfcc(y=a, sr=sr, n_mfcc=13)
    d1   = librosa.feature.delta(mfcc)
    d2   = librosa.feature.delta(mfcc, order=2)
    return np.hstack([np.mean(mfcc.T,0), np.mean(d1.T,0), np.mean(d2.T,0),
        np.mean(librosa.feature.spectral_centroid(y=a,sr=sr)),
        np.mean(librosa.feature.spectral_rolloff(y=a,sr=sr)),
        np.mean(librosa.feature.spectral_bandwidth(y=a,sr=sr)),
        np.mean(librosa.feature.zero_crossing_rate(a)),
        np.mean(librosa.feature.rms(y=a))])

# ── Sidebar filters ──────────────────────────────────────────
st.sidebar.header("Settings")
thresh    = st.sidebar.slider("Silence threshold", 0.001, 0.05, 0.01, 0.001)
rec_hours = st.sidebar.number_input("Recording duration (hours)", 1, 12, 8)

# ── File uploader ────────────────────────────────────────────
uploaded = st.file_uploader("Upload audio (.wav / .mp3)",
                             type=['wav', 'mp3', 'ogg'])

if uploaded:
    tmp = f"/tmp/{uploaded.name}"
    with open(tmp, "wb") as f: f.write(uploaded.read())

    audio, sr = preprocess_audio(tmp)

    # ── Row 1: waveform + spectrogram ────────────────────────
    col1, col2 = st.columns(2)
    with col1:
        st.subheader("Waveform")
        fig, ax = plt.subplots(figsize=(6,2))
        librosa.display.waveshow(audio, sr=sr, ax=ax, color='steelblue')
        ax.set_title(""); st.pyplot(fig); plt.close()

    with col2:
        st.subheader("Mel Spectrogram")
        mel = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=64)
        mel_db = librosa.power_to_db(mel, ref=np.max)
        fig, ax = plt.subplots(figsize=(6,2))
        img = librosa.display.specshow(mel_db, sr=sr, x_axis='time',
                                       y_axis='mel', ax=ax, cmap='magma')
        plt.colorbar(img, ax=ax, format="%+2.0f dB")
        st.pyplot(fig); plt.close()

    # ── Audio playback ───────────────────────────────────────
    st.audio(tmp)

    # ── Predict ──────────────────────────────────────────────
    try:
        with open('models.pkl', 'rb') as f:
            saved = pickle.load(f)
        svm_m, reg_m, sc = saved['svm'], saved['reg'], saved['scaler']

        feat   = extract_features(tmp)
        feat_s = sc.transform([feat])
        cls    = svm_m.predict(feat_s)[0]
        sev    = reg_m.predict(feat_s)[0]

        levels = {"No Snoring":"🟢","Mild":"🟡","Moderate":"🟠","High":"🔴","Severe":"🚨"}
        lvl_name = ("No Snoring" if sev<20 else
                    "Mild" if sev<40 else
                    "Moderate" if sev<60 else
                    "High" if sev<80 else "Severe")

        # ── Row 2: metric cards ──────────────────────────────
        st.subheader("Results")
        c1, c2, c3, c4 = st.columns(4)
        c1.metric("Detection",   "Snoring" if cls else "Non-Snoring")
        c2.metric("Severity Score", f"{sev:.1f} / 100")
        c3.metric("Level", f"{levels[lvl_name]} {lvl_name}")

        ahi = 1 / rec_hours if cls == 1 else 0
        risk_lbl = ("Low" if ahi<5 else
                    "Moderate" if ahi<15 else
                    "High" if ahi<30 else "Severe")
        c4.metric("OSA Risk", risk_lbl)

    except FileNotFoundError:
        st.warning("models.pkl not found. Run the notebook first and save models.")

else:
    st.info("👆 Upload an audio file above to begin analysis")
```
